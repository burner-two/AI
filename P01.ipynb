````python
import queue as Q  # Importing the queue module as Q to manage the city queue for BFS

# Defining the starting city and goal city for the BFS traversal
start = "Arad"  # The city where the search begins
goal = "Bucharest"  # The target city we want to reach

result = ' '  # A string to store the order of cities visited during BFS

def BFS(city, cityq, visitedq):
    """
    Perform a Breadth-First Search (BFS) to find a path from start to goal.
    
    Parameters:
    - city: The current city being explored (string).
    - cityq: The queue of cities to explore next (Q.Queue).
    - visitedq: The queue of visited cities (Q.Queue).
    """
    
    global result  # Declare result as global to modify the outer scope variable

    # If the current city is the starting city, add it to the result
    if city == start:
        result = result + " " + city
    
    # Loop through each neighboring city of the current city
    for eachcity in dict_gn[city].keys():
        # Check if the current city is the goal city
        if eachcity == goal:
            result = result + " " + eachcity  # Add goal city to the result
            return  # Exit the function since goal is found
        
        # Check if eachcity has not already been queued for exploration
        # and has not been visited yet
        if eachcity not in cityq.queue and eachcity not in visitedq.queue:
            cityq.put(eachcity)  # Add the city to the queue for future exploration
            result = result + " " + eachcity  # Add the city to the result string
    
    visitedq.put(city)  # Mark the current city as visited by adding it to the visited queue
    
    # Recursively call BFS for the next city in the queue
    BFS(cityq.get(), cityq, visitedq)

def main():
    # Initialize two queues: one for cities to explore (cityq) and one for visited cities (visitedq)
    cityq = Q.Queue()  # Queue to hold cities that need to be explored
    visitedq = Q.Queue()  # Queue to hold cities that have been visited
    
    # Start the BFS traversal from the starting city
    BFS(start, cityq, visitedq)
    
    # Output the result of the BFS traversal
    print("BFS Traversal From", start, "to", goal, "is:")
    print(result)

# Execute the main function to start the program
main()    
````

# Concept Overview

## 1. Graph Traversal
- **Graph traversal** is the process of visiting all the nodes (or vertices) in a graph. This is fundamental for many algorithms in computer science, particularly those that deal with searching and pathfinding.
- Two common methods for graph traversal are:
  - **Breadth-First Search (BFS)**: Explores all neighbors at the present depth before moving on to nodes at the next depth level. It uses a queue data structure.
  - **Depth-First Search (DFS)**: Explores as far as possible along a branch before backtracking. This can be implemented iteratively or recursively.

## 2. Search Algorithms
- **BFS** is used to find the shortest path in unweighted graphs. 
- **DFS** (and its iterative version, IDDFS) is useful for exploring all possible paths and is particularly good when the solution is far from the root.

# Depth Limited Search (DLS)

Depth-Limited Search (DLS) is a variant of the depth-first search (DFS) algorithm. It introduces a limitation on the depth of the search, preventing the algorithm from exploring paths beyond a certain depth. This can be particularly useful in situations where the search space is large, or where you want to avoid going too deep into certain paths that are unlikely to lead to a solution. Here’s a detailed breakdown of the concept:

### Key Concepts of Depth-Limited Search (DLS)

1. **Search Strategy**:
   - DLS explores a tree or graph by starting at the root node and moving downwards, following one branch until it either finds the goal or reaches the specified depth limit.
   - If the depth limit is reached and the goal has not been found, the algorithm backtracks and explores other branches.

2. **Depth Limit**:
   - The depth limit is defined as a maximum number of edges (or levels) that can be traversed from the starting node.
   - This limit helps prevent infinite loops in cases where cycles are present in the graph and restricts the search to a manageable portion of the search space.

3. **Goal Condition**:
   - As in any search algorithm, DLS checks whether the current node is the goal.
   - If the goal is found within the depth limit, the search is successful, and the algorithm can return the path taken to reach the goal.

4. **Visited Nodes**:
   - DLS maintains a list (or stack) of visited nodes to avoid revisiting them, preventing infinite loops and reducing the time complexity in scenarios with cycles.

5. **Backtracking**:
   - If the algorithm reaches the depth limit without finding the goal, it backtracks to the previous node and explores other unvisited branches.
   - This process continues until all possible paths within the depth limit are exhausted.
